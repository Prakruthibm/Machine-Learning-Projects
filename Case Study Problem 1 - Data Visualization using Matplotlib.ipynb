### 1. Sam has to build a bar-plot for the ‘Contract’ column
### a. Set the x-axis label to be ‘Contract Type of customer’
### b. Set the y-axis label to be ‘Count’
### c. Set the title of the plot to be ‘Distribution of Contract’
### d. Assign ‘orange’ color to all the bars

```python
import matplotlib.pyplot as plt
import pandas as pd


# Count the occurrences of each contract type
contract_counts = data['Contract'].value_counts()

# Create the bar plot
plt.bar(contract_counts.index, contract_counts, color='orange')

# Set the x-axis label
plt.xlabel('Contract Type of Customer')

# Set the y-axis label
plt.ylabel('Count')

# Set the title
plt.title('Distribution of Contract')

# Show the plot
plt.show()
```

### 2. Sam has to build a histogram for the ‘MonthlyCharges’ column

### 4. Sam has to build a box-plot between ‘MonthlyCharges’ & ‘PaymentMethod’. ‘MonthlyCharges’ should be on the y-axis and ‘PaymentMethod’ should be on the x-axis
### a. Set the x-axis label to be ‘Payment Method of customer’
### b. Set the y-axis label to be ‘Monthly ChargesIncurred’
### c. Set the title of plot to be ‘Monthly Charges vs. Payment Method’
### d. Assign ‘olive’ color to the box-plots

```python
data = pd.DataFrame({'MonthlyCharges': [50, 60, 70, 80, 90, 100, 110, 120],
                     'PaymentMethod': ['Credit Card', 'Electronic Check', 'Credit Card', 'Bank Transfer', 
                                       'Bank Transfer', 'Electronic Check', 'Credit Card', 'Electronic Check']})

# Create a dictionary to map colors to PaymentMethod categories
color_map = {'Credit Card': 'olive', 'Electronic Check': 'olive', 'Bank Transfer': 'olive'}

# Create a list to store MonthlyCharges data for each PaymentMethod category
payment_methods = data['PaymentMethod'].unique()
monthly_charges_data = [data[data['PaymentMethod'] == pm]['MonthlyCharges'] for pm in payment_methods]

# Create the box plot
plt.boxplot(monthly_charges_data, labels=payment_methods, patch_artist=True,
            boxprops=dict(facecolor='olive', color='black'))

# Set the x-axis label
plt.xlabel('Payment Method of Customer')

# Set the y-axis label
plt.ylabel('Monthly Charges Incurred')

# Set the title
plt.title('Monthly Charges vs Payment Method')

# Show the plot
plt.show()
```
